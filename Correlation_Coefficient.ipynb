```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
'Gender':['Male','Female','Male','Female','Male','Female','Male','Female','Male','Female',
          'Male','Female','Male','Female','Male','Female','Male','Female','Male','Female'],

'Study_Method':['Online','Offline','Online','Offline','Online','Offline','Offline','Online','Offline','Offline',
                'Online','Offline','Online','Offline','Online','Offline','Offline','Online','Offline','Offline'],

'Study_Hours':[2,5,3,6,4,7,5,4,6,8,2,7,3,6,4,7,5,4,6,8],

'Exam_Result':['Fail','Pass','Fail','Pass','Fail','Pass','Pass','Fail','Pass','Pass',
               'Fail','Pass','Fail','Pass','Fail','Pass','Pass','Fail','Pass','Pass']
}
```

```python
data= pd.DataFrame(data)
data.head(2)
```

```python
#Encoding by using the map technique
data["Gender"]=data["Gender"].map({"Male":0,"Female":1})
data["Study_Method"]=data["Study_Method"].map({"Online":0,"Offline":1})
data["Exam_Result"]=data["Exam_Result"].map({"Fail":0,"Pass":1})
```

```python
data.head(5)
```

```python
correlation = data.corr()
print(correlation)
```

```python
# # final conclusion 
# #  Study_Method → strongest relationship
#  Study_Hours → second strongest
#   Gender → weaker relationship 
```
