```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data=pd.read_csv("wine_data.csv",header=None,usecols=[11,2])
data.columns =['Alcohol','Malic acid']
```

```python
data.head(8)
```

```python
sns.pairplot(data)
```

```python
data.describe().round(2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler_df = scaler.fit_transform(data[['Alcohol', 'Malic acid']])
```

```python
scaler_df = pd.DataFrame(scaler_df,columns=['Alcohol','Malic acid'])
```

```python
sns.pairplot(scaler_df)
```

```python
scaler_df.describe().round(2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler_df = scaler.fit_transform(data[['Alcohol', 'Malic acid']])
```

```python
scaler_df = pd.DataFrame(scaler_df,columns=['Alcohol','Malic acid'])
```

```python
sns.pairplot(scaler_df)
```

```python
scaler_df.describe().round(2)
```
