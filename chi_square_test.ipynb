```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
'Study_Method':['Online','Online','Online','Online','Offline','Offline','Offline','Offline','Online','Offline',
'Online','Offline','Online','Offline','Online','Offline','Online','Offline','Online','Offline'],

'Study_Hours':['Low','Low','Medium','High','High','Medium','Low','Medium','Low','High',
'Medium','Low','High','Medium','Low','High','Medium','Low','High','Medium'],

'Attendance':['Low','Medium','Medium','High','High','High','Medium','High','Low','High',
'Medium','Low','Medium','Medium','Low','High','Medium','Low','High','High'],

'Exam_Result':['Fail','Fail','Pass','Pass','Pass','Pass','Fail','Pass','Fail','Pass',
'Pass','Fail','Pass','Pass','Fail','Pass','Pass','Fail','Pass','Pass']
}
```

```python
data = pd.DataFrame(data)
```

```python
le = LabelEncoder()
for coloum in data.columns:
      data[coloum] = le.fit_transform(data[coloum])
```

```python
x= data[["Study_Method","Study_Hours","Attendance"]]
y=data["Exam_Result"]
```

```python
chi_scores, p_values = chi2(x, y)
```

```python
print("Chi Square Scores:",chi_scores)
print("P-values:",chi_scores)
```

```python
chi_square_results = pd.DataFrame({"Feature": x.columns, "Chi Square Score": chi_scores, "P-value": p_values}) 

print(chi_square_results)
```
