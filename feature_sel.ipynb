```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load dataset
data = load_breast_cancer()

# Convert to DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target


```

```python
data.feature_names
```

```python
df.head(2)
```

```python
x=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_normalized=scaler.fit_transform(x)
# print(x_normalized)
print(x_normalized[:5])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_standardized = scaler.fit_transform(x)
print(x_standardized[:5])
```

```python
x_norm=pd.DataFrame(x_normalized,columns=x.columns)
x_norm.describe().round(3)
```

```python
x_stand = pd.DataFrame(x_standardized,columns=x.columns)
x_stand.describe().round(3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation = df.corr()
print(correlation)
```

```python
plt.Figure(figsize=(12,8))
sns.heatmap(correlation,cmap='coolwarm')
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, chi2

chi_selector = SelectKBest(chi2, k=5)
X_chi = chi_selector.fit_transform(x_normalized, y)

selected_features = x.columns[chi_selector.get_support()]
print(selected_features)
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=5000)

sfs_forward = SequentialFeatureSelector(
    model,
    n_features_to_select=5,
    direction='forward'
)

sfs_forward.fit(x_standardized, y)

selected_features = x.columns[sfs_forward.get_support()]
print(selected_features)
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=5000)

sfs_forward = SequentialFeatureSelector(
    model,
    n_features_to_select=5,
    direction='forward'
)

sfs_forward.fit(x_standardized, y)

selected_features = x.columns[sfs_forward.get_support()]
print(selected_features)
```

```python
sfs_backward = SequentialFeatureSelector(
    model,
    n_features_to_select=5,
    direction='backward'
)

sfs_backward.fit(x_standardized, y)

selected_features = x.columns[sfs_backward.get_support()]
print(selected_features)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.01)
lasso.fit(x_standardized, y)

selected_features = x.columns[lasso.coef_ != 0]
print(selected_features)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x, y)

importance = pd.Series(dt.feature_importances_, index=x.columns)
important_features = importance.sort_values(ascending=False).head(5)

print(important_features)
```
