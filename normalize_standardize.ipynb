```python
#scaling in loan dataset
```

```python
data_nor_stand=pd.read_csv("loan-test.csv")
data_nor_stand.head()
```

```python
data_nor_stand.describe()
```

```python
data_nor_stand.isnull().sum()
```

```python
sns.distplot(data_nor_stand['ApplicantIncome'])
plt.show()
```

```python
scaler1 = StandardScaler()
scaler1.fit(data_nor_stand[["ApplicantIncome"]])
```

```python
standard_coloum=scaler1.transform(data_nor_stand[["ApplicantIncome"]])
```

```python
data_nor_stand["stand_col_application"]=standard_coloum
data_nor_stand["stand_col_application"].mean()
data_nor_stand["stand_col_application"].std()
data_nor_stand.head(2)
```

```python
#observe change in data after standardscallling 
plt.Figure(figsize=(16,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data_nor_stand['ApplicantIncome'])


plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data_nor_stand["stand_col_application"])

plt.tight_layout()
plt.show()

```
