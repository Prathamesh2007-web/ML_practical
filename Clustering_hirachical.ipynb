```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\Asus\Downloads\seeds.csv")

data.head(5)     #Features: Area, Perimeter, Compactness, Length, Width, etc. This dataset contains geometrical properties of wheat kernels belonging to three different varieties of wheat.
```

```python
sns.pairplot(data, hue='V8')
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.Figure(figsize=(50,70))
sch.dendrogram(sch.linkage(data,method="single",metric='euclidean'))
plt.show()
```

```python
plt.Figure(figsize=(50,70))
sch.dendrogram(sch.linkage(data,method="ward",metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=3, linkage='single')
data['Predicted_cluster']=ac.fit_predict(data)
```
