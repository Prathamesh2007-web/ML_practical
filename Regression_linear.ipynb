```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns   
```

```python
data= pd.read_csv(r"C:\Users\Asus\Downloads\Student_Performance.csv")

data.drop("Extracurricular Activities", axis=1, inplace=True)
data.head(5)
```

```python
data.isnull().sum() 
```

```python
inp_data= data[['Hours Studied', 'Previous Scores',  'Sleep Hours', 'Sample Question Papers Practiced']]
out_data= data['Performance Index'] 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inp_data,out_data,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.score(x_train,y_train)*100
```
