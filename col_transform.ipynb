```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler,OneHotEncoder ,OrdinalEncoder
```

```python
data = pd.read_csv("col_tranform.csv")
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
x=data.drop("Passed",axis=1)
y = data["Passed"]
```

```python
numeric_coloum =["Age","StudyHours","Attendance"]
categorical_coloum=["Gender","Stream"]
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num",StandardScaler(),numeric_coloum),
        ("cat",OneHotEncoder(),categorical_coloum)
    ]
)
#syntax = transformername,transformerObject,list['coloums names on which coloum we want to apply']
```

```python
x_transformed=preprocessor.fit_transform(x)
```

```python
print(x_transformed.round(2))
print(x_transformed.shape)
```

```python
from sklearn.impute import SimpleImputer
```

```python
data_covid = pd.read_csv("covid_toy.csv")
```

```python
data_covid.head(5)
```

```python
data_covid.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test,y_train,y_test = train_test_split(data_covid.drop(columns=["has_covid"]),data_covid["has_covid"],test_size=0.2)
```

```python
X_train
```

```python
si = SimpleImputer()  # fill all the data
X_train_fever =si.fit_transform(X_train[["fever"]])
X_test_fever = si.fit_transform(X_test[["fever"]])

X_train_fever.shape
```

```python
X_train_fever
```

```python
oe= OrdinalEncoder(categories=[["Mild","Strong"]])  # apply the ordina encoding on cough coloum
x_train_cough =oe.fit_transform(X_train[["cough"]])
x_test_cough =oe.fit_transform(X_test[["cough"]])

x_train_cough.shape
```

```python
ohe = OneHotEncoder(drop="first",sparse_output=False)
x_train_gender_city = ohe.fit_transform(X_train[["gender","city"]])

x_test_gender_city=ohe.fit_transform(X_test[["gender","city"]])

x_train_gender_city.shape
```

```python
x_train_gender_city
```

```python
#for age coloum 
x_train_age = X_train.drop(columns=["gender","fever","cough","city"]).values

x_test_age = X_test.drop(columns=["gender","fever","cough","city"]).values

x_train_age.shape
```

```python
X_train_tranformed=np.concatenate((x_train_age,X_train_fever,x_train_gender_city,x_train_cough),axis=1)

X_test_tranformed=np.concatenate((x_test_age,X_test_fever,x_test_gender_city,x_test_cough),axis=1)

X_train_tranformed.shape,X_test_tranformed.shape
```
