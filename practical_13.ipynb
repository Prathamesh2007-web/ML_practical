```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame({
    'TV_Ads': np.random.randint(10, 100, 100),
    'Radio_Ads': np.random.randint(5, 50, 100),
    'Sales': np.random.randint(50, 200, 100),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 100)
})
```

```python
data['Sales'] = data['TV_Ads'] * 1.5 + data['Radio_Ads'] * 0.8 + np.random.normal(0, 10, 100)
```

```python
plt.figure(figsize=(15, 10))
plt.subplots_adjust(hspace=0.4, wspace=0.3)
```

```python
plt.subplot(2, 3, 1)
sns.histplot(data['Sales'], kde=True, color='skyblue')
plt.title('1. Histogram: Sales Distribution')
```

```python
plt.subplot(2, 3, 2)
sns.scatterplot(x='TV_Ads', y='Sales', data=data, color='green')
plt.title('2. Scatter: TV Ads vs Sales')
```

```python
plt.subplot(2, 3, 3)
sns.boxplot(x='Region', y='Sales', data=data, palette='Set2')
plt.title('3. Box Plot: Sales by Region')
```

```python
plt.subplot(2, 3, 4)
# We only calculate correlation for numeric columns
corr = data[['TV_Ads', 'Radio_Ads', 'Sales']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('4. Heatmap: Correlations')
```

```python
plt.subplot(2, 3, 5)
# We sort by TV_Ads just to show a "trend" line clearly
sorted_data = data.sort_values('TV_Ads')
plt.plot(sorted_data['TV_Ads'], sorted_data['Sales'], marker='o', color='red', linewidth=1, markersize=2)
plt.title('5. Line Chart: Sales Trend')
```
